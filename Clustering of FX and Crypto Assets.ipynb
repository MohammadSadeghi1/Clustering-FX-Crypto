### Annualized Return and Standard Deviation

```python
td_year = log_returns.count() / ((log_returns.index[-1] - log_returns.index[0]).days / 365.25)  # trading days per year
td_year
```

```python
ann_ret = mu * td_year
ann_ret
```

```python
np.exp(ann_ret) - 1       # CAGR (Compound Annual Growth Rate)
```

```python
ann_std = sigma * np.sqrt(td_year)
ann_std
```

```python
output = pd.DataFrame(data={'Ann_Return': ann_ret, 'Ann_Risk': ann_std})
output
```

```python
output.rank(ascending=False).sort_values('Ann_Return') 
```

```python
r_f = 0       #  risk free rate (simplification)
```

```python
def sharpe(log_ret, r_f=0):
    
    if log_ret.std() == 0:
        return np.nan
    else:
        return(log_ret.mean() - r_f) / log_ret.std() * np.sqrt(log_ret.count() / ((log_ret.index[-1] - log_ret.index[0]).days / 365.25))
```

```python
log_returns.apply(sharpe).sort_values(ascending=False)
```

```python
output['Sharpe'] = log_returns.apply(sharpe)
output
```

## Sortino Ratio

```python
def sortino(log_ret, r_f=0):       
    excess_returns = log_ret - r_f
    excess_returns = np.where(excess_returns<0, excess_returns, 0)
    downside_deviation = np.sqrt(np.mean(excess_returns ** 2))
    if downside_deviation == 0:
        return np.nan
    else:
        return ((log_ret.mean() - r_f) / downside_deviation) * np.sqrt(log_ret.count() / ((log_ret.index[-1] - log_ret.index[0]).days / 365.25))
```

```python
log_returns.apply(sortino).sort_values(ascending=False)
```

```python
output ['Sortino'] = log_returns.apply(sortino)
output
```

## Calmar Ratio

```python
df
```

```python
BTC = df['BTC-USD'].to_frame()
BTC
```

```python
BTC.rename(columns={'BTC-USD': 'Norm_Price'}, inplace=True)
BTC
```

```python
BTC['Cummax'] = BTC['Norm_Price'].cummax()
BTC
```

```python
BTC['Drawdown'] = (BTC['Cummax'] - BTC['Norm_Price']) / BTC['Cummax']      # (pos.) drawdown in %      
BTC
```

```python
BTC['Drawdown'].max()
```

```python
BTC.plot(figsize=(15,10), fontsize=13, secondary_y='Drawdown') 
plt.title('BTC', fontdict={'fontweight': 'bold', 'fontsize': 18})
plt.legend(fontsize=13)
plt.show()
```

```python
def max_drawdown(log_ret):
    norm_prices = log_ret.cumsum().apply(np.exp)
    cummax = norm_prices.cummax()
    drawdown = (cummax - norm_prices) / cummax
    max_dd = drawdown.max()
    return max_dd
```

```python
log_returns.apply(max_drawdown).sort_values(ascending=False)
```

```python
def calmar(log_ret):
    max_dd = max_drawdown(log_ret)
    if max_dd == 0:
        return np.nan
    else:
        cagr = np.exp(log_ret.sum()) ** (1 / ((log_ret.index[-1] - log_ret.index[0]).days / 365.25)) - 1
        return cagr / max_dd
```

```python
log_returns.apply(calmar).sort_values(ascending=False)
```

```python
output['Calmar'] = log_returns.apply(calmar)
output.sort_values('Sharpe', ascending=False)
```

```python
df = output.rank(ascending=False).sort_values('Sharpe')
df
```
